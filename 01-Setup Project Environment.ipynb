1. create external location gunjanadls2
2. create catalog formula1
3. create schema landing bronze, silver & gold
4. create volume file in the landing schema

```sql
%fs ls 'abfss://formula1@gunjanadlsgen2.dfs.core.windows.net/landing'
```

```sql
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databricks_ext_loc_demo
URL 'abfss://formula1@gunjanadlsgen2.dfs.core.windows.net'
WITH (STORAGE CREDENTIAL `dbt_credential`)
COMMENT 'external location for formula1 container';
```

```sql
%sql 
show catalogs
```

```sql
%sql
create catalog if not exists formula1
    managed location 'abfss://formula1@gunjanadlsgen2.dfs.core.windows.net'
    comment 'this is the main catalog for formula1 project'
```

- Create SCHEMA BRONZE, SILVER AND GOLD

```sql
CREATE SCHEMA IF NOT EXISTS formula1.landing;
CREATE SCHEMA IF NOT EXISTS formula1.bronze
   managed location 'abfss://formula1@gunjanadlsgen2.dfs.core.windows.net';
CREATE SCHEMA IF NOT EXISTS formula1.silver
   managed location 'abfss://formula1@gunjanadlsgen2.dfs.core.windows.net';
CREATE SCHEMA IF NOT EXISTS formula1.gold
   managed location 'abfss://formula1@gunjanadlsgen2.dfs.core.windows.net';
```

```sql
use catalog formula1;
show schemas
```

```sql
CREATE EXTERNAL VOLUME formula1.landing.files
LOCATION 'abfss://formula1@gunjanadlsgen2.dfs.core.windows.net/landing';
```
