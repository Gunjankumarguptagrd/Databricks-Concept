```python
item_price = 120
tax_rate = 20 # in percent
discount = 10
# calculate tax to applied
tax = item_price * tax_rate / 100
# calculate discounted price
discounted_price_with_tax = item_price + tax
# calculate final price
final_price = discounted_price_with_tax - discount
print(f"final price is : {final_price}")
```
